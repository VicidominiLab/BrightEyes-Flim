### Formulas DFD ###
$$\tau_{\phi} = {{1}\over{2\pi f}} tan(\phi)$$

$$\tau_m = {{1}\over{2\pi f}} \sqrt{{{1\over{m^2}} -1 }}$$

$$m = \sqrt{{1}\over{  {\left(  {tan \phi}  \right)}^2 + 1}} = \sqrt{cos^2 \phi} = | cos \phi | $$  

$$\phi  = arctan \left( 2 \pi {{\tau}\over{t_{DFD}}} \right)$$  

$$\rho = 0.5^2 + 0.5 \left( 1 - sin \theta \right)$$
